```python
Machine learning is basically a way of teaching computers to learn from experience,
just like humans do. Instead of telling the computer every single rule step-by-step, 
we give it a lot of examples along with the correct answers, and it learns the pattern on its own.
Over time, it improves its understanding and can make predictions or decisions when it sees new data. 
This is different from traditional programming, where we manually write all the rules. 
Machine Learning is useful when problems are too complex to solve with fixed rules, 
such as recognizing faces, predicting prices, or detecting spam emails
```

```python
##Supervised Learning
```

```python
Supervised Learning is a type of Machine Learning where the data comes with labels (correct answers).
The model learns the relationship between inputs and outputs so it can predict the correct answer for new, unseen data.

Example: If emails are labeled as “spam” or “not spam,” the model learns to classify future emails correctly.
```

```python
###Unsupervised Learning
```

```python
Unsupervised Learning is used when the data has no labels. The model tries to find hidden patterns, structures, 
or clusters within the data on its own.

Example: Grouping customers into segments based on their buying behavior without predefined categories.
```

```python
##Reinforcement Learning
```

```python
Reinforcement Learning is a type of learning where an agent learns by interacting with an environment. 
It receives rewards for correct actions and penalties for wrong actions, and over time it learns the best strategy to maximize total reward.

Example: A game-playing AI improving its moves based on wins and losses.
```

```python
##Comparison table
```

```python
| Learning Type          | Data Has Labels? | How It Learns                                     | Real-World Examples                           |
| ---------------------- | ---------------- | ------------------------------------------------- | --------------------------------------------- |
| Supervised Learning    | ✅ Yes            | Learns from input-output pairs                    | Spam Detection, House Price Prediction        |
| Unsupervised Learning  | ❌ No             | Finds hidden patterns/clusters                    | Customer Segmentation, Market Basket Analysis |
| Reinforcement Learning | Reward-based     | Learns via trial and error with rewards/penalties | Self-Driving Cars, Game Playing AI            |
```
